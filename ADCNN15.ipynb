```python
# deblur_train_full.py
# Notebook-friendly, pairs files smartly (name normalization) with index fallback.
# Default: 100 epochs + periodic checkpoints. 15-layer DeblurCNN.

import os, time, argparse, re, random
from pathlib import Path
from collections import defaultdict

import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import transforms
from torchvision.utils import save_image
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import matplotlib.pyplot as plt

# --------------------------
# Reproducibility
# --------------------------
def set_seed(seed: int = 108):
    random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(108)
torch.backends.cudnn.benchmark = True  # speed on constant input sizes

# --------------------------
# Args (safe for Jupyter)
# --------------------------
parser = argparse.ArgumentParser()
parser.add_argument("-f", "--fff", default="1")                 # dummy for ipykernel
parser.add_argument("--epochs", type=int, default=100)
parser.add_argument("--batch_size", type=int, default=4)
parser.add_argument("--lr", type=float, default=1e-4)
parser.add_argument("--img_size", type=int, default=224)
parser.add_argument("--blur_dir", type=str, default="./blur")
parser.add_argument("--sharp_dir", type=str, default="./deblur")
parser.add_argument("--out_dir", type=str, default="./outputs")
parser.add_argument("--ckpt_every", type=int, default=10)       # checkpoint frequency (epochs)
args = vars(parser.parse_args())

OUT_DIR   = Path(args["out_dir"])
SAVED_DIR = OUT_DIR / "saved_images"
SAVED_DIR.mkdir(parents=True, exist_ok=True)

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# --------------------------
# File pairing helpers
# --------------------------
IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff", ".webp"}

def list_images(root: str):
    root = Path(root)
    return [p for p in root.rglob("*") if p.suffix.lower() in IMG_EXTS and p.is_file()]

def normalize_stem(p: Path) -> str:
    s = p.stem.lower()
    tokens = [
        "deblur", "unblur", "sharp", "clear", "target", "label",
        "groundtruth", "ground_truth", "gt", "truth",
        "blur", "blurry", "input", "source"
    ]
    for t in tokens:
        s = s.replace(t, "")
    s = re.sub(r"^(img|image|frame|frm|pic|photo)+", "", s)
    s = re.sub(r"[_\-\s]+", "", s)
    s = s.strip()
    return s

def pair_smart(blur_paths, sharp_paths):
    bm = defaultdict(list)
    sm = defaultdict(list)
    for p in blur_paths:  bm[normalize_stem(p)].append(p)
    for p in sharp_paths: sm[normalize_stem(p)].append(p)

    common_keys = sorted(set(bm.keys()) & set(sm.keys()))
    blur_paired, sharp_paired = [], []
    for k in common_keys:
        b_list = sorted(bm[k])
        s_list = sorted(sm[k])
        m = min(len(b_list), len(s_list))
        blur_paired.extend(b_list[:m])
        sharp_paired.extend(s_list[:m])

    if len(blur_paired) == 0:
        blur_sorted  = sorted(blur_paths)
        sharp_sorted = sorted(sharp_paths)
        m = min(len(blur_sorted), len(sharp_sorted))
        blur_paired  = blur_sorted[:m]
        sharp_paired = sharp_sorted[:m]
        print(f"[WARN] No normalized-stem matches found. Falling back to index-based pairing for {m} pairs.")

    return blur_paired, sharp_paired

# --------------------------
# Discover & pair data
# --------------------------
blur_dir  = args["blur_dir"]
sharp_dir = args["sharp_dir"]

all_blur  = list_images(blur_dir)
all_sharp = list_images(sharp_dir)

if len(all_blur) == 0 or len(all_sharp) == 0:
    raise RuntimeError(f"No images found. blur_dir='{blur_dir}' ({len(all_blur)}) | sharp_dir='{sharp_dir}' ({len(all_sharp)}).")

blur_paired, sharp_paired = pair_smart(all_blur, all_sharp)

print(f"Total paired images: {len(blur_paired)}")
for i in range(min(3, len(blur_paired))):
    print("  BLUR  →", blur_paired[i].name, " | SHARP →", sharp_paired[i].name)

x_train, x_val, y_train, y_val = train_test_split(
    blur_paired, sharp_paired, test_size=0.25, random_state=42, shuffle=True
)
print(f"Train: {len(x_train)} | Val: {len(x_val)}")

# --------------------------
# Transforms
# --------------------------
IMG_SIZE = args["img_size"]
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

# --------------------------
# Dataset
# --------------------------
class DeblurDataset(Dataset):
    def __init__(self, blur_paths, sharp_paths=None, transforms=None):
        self.X = list(blur_paths)
        self.y = list(sharp_paths) if sharp_paths is not None else None
        self.transforms = transforms
        if self.y is not None and len(self.X) != len(self.y):
            m = min(len(self.X), len(self.y))
            self.X = self.X[:m]
            self.y = self.y[:m]

    def __len__(self):
        return len(self.X)

    def _read_rgb(self, path: Path):
        img = cv2.imread(str(path), cv2.IMREAD_COLOR)
        if img is None:
            raise FileNotFoundError(f"Could not read image: {path}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return img

    def __getitem__(self, idx):
        blur_img = self._read_rgb(self.X[idx])
        if self.transforms:
            blur_img = self.transforms(blur_img)
        if self.y is not None:
            sharp_img = self._read_rgb(self.y[idx])
            if self.transforms:
                sharp_img = self.transforms(sharp_img)
            return blur_img, sharp_img
        return blur_img

BATCH = args["batch_size"]
train_data = DeblurDataset(x_train, y_train, transform)
val_data   = DeblurDataset(x_val,   y_val,   transform)

trainloader = DataLoader(train_data, batch_size=BATCH, shuffle=True,  num_workers=0,
                         pin_memory=torch.cuda.is_available())
valloader   = DataLoader(val_data,   batch_size=BATCH, shuffle=False, num_workers=0,
                         pin_memory=torch.cuda.is_available())

# --------------------------
# Model (15-layer)
# --------------------------
class DeblurCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1  = nn.Conv2d(3,   64, kernel_size=9, padding=9//2)
        self.conv2  = nn.Conv2d(64,  64, kernel_size=3, padding=1)
        self.conv3  = nn.Conv2d(64,  64, kernel_size=3, padding=1)
        self.conv4  = nn.Conv2d(64,  64, kernel_size=3, padding=1)
        self.conv5  = nn.Conv2d(64,  64, kernel_size=3, padding=1)

        self.conv6  = nn.Conv2d(64,  32, kernel_size=1, padding=0)

        self.conv7  = nn.Conv2d(32,  32, kernel_size=3, padding=1)
        self.conv8  = nn.Conv2d(32,  32, kernel_size=3, padding=1)
        self.conv9  = nn.Conv2d(32,  32, kernel_size=3, padding=1)
        self.conv10 = nn.Conv2d(32,  32, kernel_size=3, padding=1)

        self.conv11 = nn.Conv2d(32,  16, kernel_size=1, padding=0)

        self.conv12 = nn.Conv2d(16,  16, kernel_size=3, padding=1)
        self.conv13 = nn.Conv2d(16,  16, kernel_size=3, padding=1)
        self.conv14 = nn.Conv2d(16,  16, kernel_size=3, padding=1)
        self.conv15 = nn.Conv2d(16,   3, kernel_size=5, padding=5//2)

        self.relu = nn.ReLU(inplace=True)

        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, nonlinearity="relu")
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.relu(self.conv3(x))
        x = self.relu(self.conv4(x))
        x = self.relu(self.conv5(x))

        x = self.relu(self.conv6(x))

        x = self.relu(self.conv7(x))
        x = self.relu(self.conv8(x))
        x = self.relu(self.conv9(x))
        x = self.relu(self.conv10(x))

        x = self.relu(self.conv11(x))

        x = self.relu(self.conv12(x))
        x = self.relu(self.conv13(x))
        x = self.relu(self.conv14(x))
        x = self.conv15(x)
        return x

model = DeblurCNN().to(device)
print(model)

# --------------------------
# Loss / Optim / Scheduler
# --------------------------
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=args["lr"])
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', patience=5, factor=0.1, verbose=True
)

# --------------------------
# Utils
# --------------------------
def save_batch_images(tensor_bchw, path: Path):
    save_image(tensor_bchw, str(path))

def save_checkpoint(model, out_dir: Path, epoch: int):
    ckpt_path = out_dir / f"model_ep{epoch}.pth"
    torch.save(model.state_dict(), ckpt_path)

# --------------------------
# Train / Validate loops
# --------------------------
def fit(model, dataloader, epoch):
    model.train()
    running = 0.0
    pbar = tqdm(dataloader, total=len(dataloader), desc=f"Train {epoch+1}")
    for blur_img, sharp_img in pbar:
        blur_img  = blur_img.to(device, non_blocking=True)
        sharp_img = sharp_img.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        out = model(blur_img)
        loss = criterion(out, sharp_img)
        loss.backward()
        optimizer.step()

        running += loss.item()
        pbar.set_postfix(loss=f"{loss.item():.4f}")
    return running / max(1, len(dataloader))

def validate(model, dataloader, epoch):
    model.eval()
    running = 0.0
    last_triplet = None
    pbar = tqdm(dataloader, total=len(dataloader), desc=f"Valid {epoch+1}")
    with torch.no_grad():
        for blur_img, sharp_img in pbar:
            blur_img  = blur_img.to(device, non_blocking=True)
            sharp_img = sharp_img.to(device, non_blocking=True)
            out = model(blur_img)
            loss = criterion(out, sharp_img)
            running += loss.item()
            last_triplet = (blur_img, sharp_img, out)
            pbar.set_postfix(loss=f"{loss.item():.4f}")

    if last_triplet is not None:
        b, s, o = last_triplet
        save_batch_images(s.cpu(), SAVED_DIR / f"sharp_ep{epoch+1}.jpg")
        save_batch_images(b.cpu(), SAVED_DIR / f"blur_ep{epoch+1}.jpg")
        save_batch_images(o.cpu(), SAVED_DIR / f"deblur_ep{epoch+1}.jpg")

    return running / max(1, len(dataloader))

# --------------------------
# Run training
# --------------------------
E = int(args["epochs"])
CKPT_EVERY = max(0, int(args["ckpt_every"]))  # 0 disables periodic ckpts

train_loss, val_loss = [], []
start = time.time()

for ep in range(E):
    print(f"\nEpoch {ep+1}/{E}")
    tl = fit(model, trainloader, ep)
    vl = validate(model, valloader, ep)
    train_loss.append(tl); val_loss.append(vl)
    print(f"Train Loss: {tl:.6f} | Val Loss: {vl:.6f}")
    scheduler.step(vl)

    if CKPT_EVERY and (ep + 1) % CKPT_EVERY == 0:
        save_checkpoint(model, OUT_DIR, ep + 1)

elapsed_min = (time.time() - start)/60.0
print(f"Took {elapsed_min:.3f} minutes to train")

# --------------------------
# Plot & save artifacts
# --------------------------
OUT_DIR.mkdir(parents=True, exist_ok=True)
plt.figure(figsize=(8,5))
plt.plot(train_loss, label='train')
plt.plot(val_loss, label='val')
plt.xlabel('epoch'); plt.ylabel('MSE loss'); plt.legend()
plt.tight_layout()
plt.savefig(OUT_DIR / "loss.png")
plt.show()

print("Saving final model...")
torch.save(model.state_dict(), OUT_DIR / "model.pth")
print("Done. Artifacts saved in:", OUT_DIR.resolve())
```
